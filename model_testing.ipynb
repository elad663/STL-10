```lua
-- Conv layer (23 channels, 7x7 filters, stride 2, padding 2, RELU activation)
-- Max pooling (3x3 patch, stride 2)
-- Dropout
-- Full connected layer (50 units)
-- Softmax layer

criterion = nn.ClassNLLCriterion()

model = nn.Sequential()
-- 3 x 96 x 96
model:add(nn.SpatialConvolution(3, 23, 7, 7, 2, 2))
-- 23 x 45 x 45
model:add(nn.ReLU())
model:add(nn.SpatialMaxPooling(3,3,2,2))
-- 23 x 22 x 22
model:add(nn.Dropout(.5))
model:add(nn.Reshape(23*22*22))
-- 11132
model:add(nn.Linear(23*22*22, 50))
-- 50
model:add(nn.Linear(50,10))
-- 10
model:add(nn.SoftMax())
```

```lua
x2 = model:forward(x)
print(#x2)
```

```lua
err = criterion:forward(x2, 10)
df_do = criterion:backward(x2, 10)
```
